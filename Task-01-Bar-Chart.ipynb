## Imporitng the Dataset of Covid-19 Golobal Dataset

### Importing Necessary Labriers

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
file_path = 'F:\\Experiences\\01-Prodigy-InfoTech-Internship\\archive_5\\worldometer_coronavirus_summary_data.csv'

# Try reading the file with some common fixes

df = pd.read_csv(file_path, delimiter=',', on_bad_lines='skip', low_memory=False)
```

```python
df.head()
```

```python
df.columns
```

```python
# Example: Bar chart of total cases by country
df.groupby('country')['total_confirmed'].sum()
```

###  Grouping the data by continent and summing up the total confirmed cases

```python

continent_cases = df.groupby('continent')['total_confirmed'].sum().reset_index()
```

## Total Confirmed COVID-19 Cases by Continent

```python
# Creating the bar chart
plt.figure(figsize=(10, 6))
plt.bar(continent_cases['continent'], continent_cases['total_confirmed'], color='skyblue')
plt.title('Total Confirmed COVID-19 Cases by Continent')
plt.xlabel('Continent')
plt.ylabel('Total Confirmed Cases')

# Displaying y-axis labels in plain numeric format
plt.ticklabel_format(style='plain', axis='y')
plt.xticks(rotation=45)

# Adding commas as thousand separators
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.show()
```

## Countries with the Most and Least Confirmed COVID-19 Total Deaths Bar Chart

```python
# Find the country with the most and least confirmed cases
most_cases_country = df.loc[df['total_deaths'].idxmax()]
least_cases_country = df.loc[df['total_deaths'].idxmin()]

# Create a DataFrame with these two countries
comparison_df = pd.DataFrame({
    'Country': [most_cases_country['country'], least_cases_country['country']],
    'Total Confirmed Cases': [most_cases_country['total_deaths'], least_cases_country['total_deaths']]
})

# Plotting the bar chart
plt.figure(figsize=(8, 6))
plt.bar(comparison_df['Country'], comparison_df['Total Confirmed Cases'], color=['green', 'red'])
plt.title('Countries with the Most and Least Confirmed COVID-19 Total Deaths')
plt.xlabel('Country')
plt.ylabel('Total Deaths')

# Display y-axis labels in plain numeric format with commas
plt.ticklabel_format(style='plain', axis='y')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.show()
```

## Top 5 Countries With Most Confirmed COVID-19 Total Deaths Bar Chart

```python
# Sort the DataFrame by total deaths in descending order
sorted_df = df.sort_values(by='total_deaths', ascending=False)

# Select the top 5 countries with the most confirmed cases
top_5_countries = sorted_df.head(5)

# Plotting the bar chart
plt.figure(figsize=(10, 8))
plt.bar(top_5_countries['country'], top_5_countries['total_deaths'], color='blue')
plt.title('Top 5 Countries with the Most Confirmed COVID-19 Total Deaths')
plt.xlabel('Country')
plt.ylabel('Total Deaths')

# Display y-axis labels in plain numeric format with commas
plt.ticklabel_format(style='plain', axis='y')
plt.gca().yaxis.set_major_formatter(plt.FuncFormatter(lambda x, _: f'{int(x):,}'))

plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.tight_layout()  # Adjust layout to fit labels
plt.show()
```
